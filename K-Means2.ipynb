```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt
%matplotlib inline
```

```python

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df = df.drop(columns=['petal length (cm)','petal width (cm)'])
df
```

```python
plt.scatter(df['sepal length (cm)'],df['sepal width (cm)'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['sepal length (cm)','sepal width (cm)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'],color='green')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'],color='red')
plt.scatter(df3['sepal length (cm)'], df3['sepal width (cm)'],color='blue')

plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['sepal width (cm)']])
df['sepal width (cm)'] = scaler.transform(df[['sepal width (cm)']])
df
```

```python
scaler.fit(df[['sepal length (cm)']])
df['sepal length (cm)'] = scaler.transform(df[['sepal length (cm)']])
df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['sepal length (cm)','sepal width (cm)']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'],color='green')
plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'],color='red')
plt.scatter(df3['sepal length (cm)'], df3['sepal width (cm)'],color='blue')

plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
```

```python
km.cluster_centers_
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['sepal length (cm)','sepal width (cm)']])
    sse.append(km.inertia_)

sse
```
